```python
stars = ["Star" + str(i) for i in range(1,11)]
dfu = df[["First.Name", "Last.Name", "Middle.Initial", "Suffix.Name", "Race", "Gender", "Current.Age", "Appointed.Date"] + stars].drop_duplicates()
```

```python
id_cols = ["First.Name", "Last.Name", "Middle.Initial", "Suffix.Name", "Appointed.Date", "Current.Age"]
```

```python
KeepDuplicates(dfu, id_cols)
```

```python
dfu = AssignUniqueIDs(dfu,id_cols)
```

```python
tids = KeepDuplicates(dfu, id_cols)['TID'].tolist()
```

```python
max_cols = stars; uid_col = ['TID']
```

```python
dfa = dfu[uid_col + max_cols].groupby(uid_col, as_index=False).agg(max)
```
