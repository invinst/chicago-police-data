```python
import pandas as pd
from assign_unique_ids_functions import *
```

```python
df = pd.read_csv("../output/unit-history.csv.gz")
```

```python
id_cols = ["TID", "First.Name", "Last.Name", "Middle.Initial", "Suffix.Name", "Appointed.Date", "Current.Age", "Race","Gender"]
```

```python
no_unit = list(df.loc[df['Unit'] == -999, 'TID'])
sometimes_no_unit = list(df.loc[(df['TID'].isin(no_unit)) & (df['Unit'] > 0), 'TID'].drop_duplicates())
df.loc[(df['TID'].isin(sometimes_no_unit)), id_cols + ['Unit', 'Effective.Date'] + ["Star1", "Star2", "Star3"]]
```

```python
never_unit = list(df.loc[(df['TID'].isin(no_unit)) & (~ df['TID'].isin(sometimes_no_unit)) , 'TID'].drop_duplicates())
df.loc[(df['TID'].isin(never_unit)), id_cols + ['Unit', 'Effective.Date'] + ["Star1", "Star2", "Star3"]]
```
