```python
id_cols = ["first_name_NS", "last_name_NS", "middle_initial", "middle_initial2", "suffix_name", "gender"]
sub_id_cols = ["first_name_NS", "last_name_NS", "middle_initial", "middle_initial2", "suffix_name", "gender", "year"]
sub_conflict_cols = ['org_hire_date', 'age_at_hire']
mid = 'salary_2002-2017_2017-09_ID'
sub_id = 'salary-year_2002-2017_2017-09_ID'
```

```python
out_df = pd.DataFrame()
agg_full_df = pd.DataFrame()
for year in df['year'].unique():
    print(year)
     #log.info('Assigning unique sub-ids for year: %d', year)
    sub_df = df[df['year']==year]
    sub_df = assign_unique_ids(sub_df, sub_id,
                                sub_id_cols,
                                conflict_cols=sub_conflict_cols,
                                log=False)
    sub_df[sub_id] = sub_df[sub_id] + year * 100000
    out_df = out_df.append(sub_df)
    sub_agg_df = aggregate_data(sub_df, sub_id,
                                 sub_id_cols,
                                 max_cols=sub_conflict_cols + ['salary'])
    agg_full_df = agg_full_df.append(sub_agg_df)
```
